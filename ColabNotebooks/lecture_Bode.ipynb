```python
import numpy as np
from scipy import signal
import matplotlib.pyplot as plt

sys = signal.TransferFunction([1], [1, 1])
array_w, mag, phase = signal.bode(sys)

plt.figure()
plt.semilogx(array_w, mag)    # Bode magnitude plot
plt.figure()
plt.semilogx(array_w, phase)  # Bode phase plot
plt.show()
```

```python
from math import atan2, pi

def my_tf_w(w):
    re =  1 / (1 + w*w)
    im = -w / (1 + w*w)
    return re, im 

Count = array_w.shape[0]
array_amp = np.zeros((Count, ))
array_arg = np.zeros((Count, ))

for i in range(Count):
    w = array_w[i]
    re, im = my_tf_w(w)
    array_amp[i] = 20*np.log10( np.sqrt(re*re + im*im) ) 
    array_arg[i] = atan2(im, re) * 180 / pi

plt.figure()
plt.semilogx(array_w, array_amp)    # Bode magnitude plot
plt.figure()
plt.semilogx(array_w, array_arg)  # Bode phase plot
plt.show()
```
